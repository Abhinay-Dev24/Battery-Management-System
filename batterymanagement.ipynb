```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import plotly.express as px

df = pd.read_csv('/content/finalbattery.csv')
print(df.columns)
df.head()
X_SOH = df[['Initial capacity (Ah)', 'EOL capacity(Ah)', 'Current(A)', 'Cut-off voltage(V)']]
Y_SOH = df['SOH']
X_SOC = df[[ 'Temperature(째C)', 'Current(A)', 'Cut-off voltage(V)']]
Y_SOC = df['SOC']
X_SOH_train, X_SOH_test, Y_SOH_train, Y_SOH_test = train_test_split(X_SOH, Y_SOH, test_size=0.2, random_state=42)
X_SOC_train, X_SOC_test, Y_SOC_train, Y_SOC_test = train_test_split(X_SOC, Y_SOC, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_SOH_train_scaled = scaler.fit_transform(X_SOH_train)
X_SOH_test_scaled = scaler.transform(X_SOH_test)
X_SOC_train_scaled = scaler.fit_transform(X_SOC_train)
X_SOC_test_scaled = scaler.transform(X_SOC_test)
svm_SOH = SVR()
svm_SOH.fit(X_SOH_train_scaled, Y_SOH_train)
svm_SOC = SVR()
svm_SOC.fit(X_SOC_train_scaled, Y_SOC_train)
Y_SOH_pred = svm_SOH.predict(X_SOH_test_scaled)
print(Y_SOH)
Y_SOC_pred = svm_SOC.predict(X_SOC_test_scaled)
print(Y_SOC)
mse_SOH = mean_squared_error(Y_SOH_test, Y_SOH_pred)
mse_SOC = mean_squared_error(Y_SOC_test, Y_SOC_pred)
print(f'Mean Squared Error for SOH prediction: {mse_SOH}')
print(f'Mean Squared Error for SOC prediction: {mse_SOC}')
soh_plot = px.scatter_3d(x=Y_SOH_test, y=X_SOH_test['Initial capacity (Ah)'], z=Y_SOH_pred,
                           labels={'x': 'Actual SOH', 'y': 'Initial capacity (Ah)', 'z': 'Predicted SOH'},
                           title='Actual vs Predicted SOH')
soc_plot = px.scatter_3d(x=Y_SOC_test, y=X_SOC_test['Temperature(째C)'], z=Y_SOC_pred,
                           labels={'x': 'Actual SOC', 'y': 'Temperature(째C)', 'z': 'Predicted SOC'},
                           title='Actual vs Predicted SOC')
soc_plot.show()
soh_plot.show()
```
